```python
from pydantic_ai import Agent
from pydantic_ai.settings import ModelSettings
from pydantic_ai.models.openai import OpenAIChatModel
from pydantic_ai.providers.openai import OpenAIProvider
from deepeval.integrations.pydantic_ai import ConfidentInstrumentationSettings

import os
from dotenv import load_dotenv

load_dotenv("/Users/phung.pham/Documents/PHUNGPX/deepeval_exploration/agentic/.env")

llm_model = os.getenv("LLM_MODEL")
llm_base_url = os.getenv("LLM_BASE_URL")
llm_api_key = os.getenv("LLM_API_KEY")
llm_temperature = os.getenv("LLM_TEMPERATURE")
llm_max_tokens = os.getenv("LLM_MAX_TOKENS")

model = OpenAIChatModel(
    model_name=llm_model,
    provider=OpenAIProvider(
        base_url=llm_base_url,
        api_key=llm_api_key,
    ),
    settings=ModelSettings(
        temperature=0,
        max_tokens=2000,
    ),
)

agent = Agent(
    model=model,
    system_prompt="Be concise, reply with one sentence.",
    instrument=ConfidentInstrumentationSettings(),
    name="test_agent",
)
```

```python
a = agent.run("What are the LLMs?")
```
